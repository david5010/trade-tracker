```python
trades_numeric = trades.select_dtypes(include=['int', 'float', 'int64', 'float64'])
```

```python
trades_numeric.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the aesthetics for better visibility
sns.set(style="whitegrid")

# Loop through the numeric columns and create a distribution plot for each
for column in trades_numeric.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(trades_numeric[column], kde=True, bins=50)  # `kde=True` adds a Kernel Density Estimate to smooth the histogram
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
